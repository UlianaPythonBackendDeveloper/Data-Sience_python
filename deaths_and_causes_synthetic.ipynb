```python
import pandas as pd
df = pd.read_csv('/content/deaths_and_causes_synthetic.csv')
# 1. Сумма смертей по стране
country_summary = df.groupby('Country')['Number_of_Deaths'].sum().sort_values(ascending=False)
print("Топ-10 стран по общему числу смертей:")
print(country_summary.head(10))
print()
# 2. Средняя смертность по причинам
cause_summary = df.groupby('Cause_of_Death')['Mortality_Rate_per_1000'].mean().sort_values(ascending=False)
print("Средний уровень смертности по причинам:")
print(cause_summary)
print()
# 3. Количество записей по годам
year_counts = df['Year'].value_counts().sort_index()
print("Количество записей по годам:")
print(year_counts)
print()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
country_summary.head(10).plot(kind='bar',color='steelblue')
plt.title('Топ-10 стран по общему числу смертей')
plt.xlabel('Страна')
plt.ylabel('Число смертей')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
cause_counts = df['Cause_of_Death'].value_counts()
cause_counts.plot(kind='pie',autopct='%1.1f%%')
plt.title('Распределение причин смерти')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
